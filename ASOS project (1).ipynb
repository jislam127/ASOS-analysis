Our next analysis, consisted of sorting each clothing item into labels, based on price. We achieved this using bins. Using bins, ranging from 50 to 300, and labels starting from affordable, and ending at luxury, we were able to sort the pieces into their own price based categories.

```python
print(df3['price.current.value'])
    
```

```python
bins = [50, 75, 125, 200, 250, 300]
labels = ['affordable','mid-budget','medium','expensive','luxury']
categories = pd.cut(df3['price.current.value'], bins, labels=labels)
print(categories.value_counts())
```

Finding the average price based on the brand, using the groupby function, we are then able to plot a line plot, highlighting the different average prices across each brand, and how each brand differ in price.

```python
import seaborn as sns
import numpy as np
```

```python
pivot_clothes = df3.pivot(index = 'brandName', columns = 'name', values = 'price.current.value')
```

```python
grouped_df = df3.groupby(['brandName'])['price.current.value'].mean().reset_index()
grouped_df
```

```python
sns.lineplot(x=df3['brandName'],y=df3['price.current.value'], data ='grouped_df', marker = 'o', markersize = 5)
plt.xticks(rotation = 90)
plt.ylabel('Price')
plt.xlabel('Brand')
plt.title('Average Prices Across Brands')
plt.show()
```

Maids to Measure, Six Stories, and Bardot are some of the most expensive brands on average.
